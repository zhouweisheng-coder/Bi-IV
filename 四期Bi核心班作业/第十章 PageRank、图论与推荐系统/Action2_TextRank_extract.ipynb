# Action2使用TextRank对新闻进行关键词提取，及文章摘要输出news.txt

1、正确读取文件 （20points）     2、使用TextRank工具进行关键词提取，结果正确 （20points）     3、使用TextRank工具进行重要段落筛选，并形成摘要（20points）

```python
#-*- encoding:utf-8 -*-
from textrank4zh import TextRank4Keyword, TextRank4Sentence
import jieba
import jieba.analyse
import jieba.posseg as pseg # 词性cut
```

```python
# 新闻读取
with open('news_高考.txt', 'r',encoding='utf-8') as f:
    text = f.read()
with open('news_土耳其.txt', 'r',encoding='utf-8') as f:
    text += f.read()
with open('news_武汉.txt', 'r',encoding='utf-8') as f:
    text += f.read()
text
```

```python
# 输出关键词，设置文本小写，窗口为5
tr4w = TextRank4Keyword()
tr4w.analyze(text=text, lower=True, window=5)
print('关键词：')
for item in tr4w.get_keywords(20, word_min_len=2):
    print(item.word, item.weight)
```

```python
# 使用jieba基于TextRank算法的关键词抽取
#keywords = jieba.analyse.extract_tags(sentence, topK=20, withWeight=True, allowPOS=('n','nr','ns'))
keywords = jieba.analyse.textrank(text, topK=20, withWeight=True) #默认allowPOS=('ns', 'n', 'vn', 'v')
print(keywords)
for item in keywords:
    print(item[0],item[1])
```

```python
# 使用jieba
seg_list = jieba.cut(text, cut_all=False)
print(' '.join(seg_list))
# 获取分词和词性
words = pseg.cut(text)
for word, flag in words:
    print('%s, %s' % (word, flag))
```

```python
# 通过TF-IDF获取关键词
keywords = jieba.analyse.extract_tags(text, topK=20, withWeight=True, allowPOS=('n','nr','ns'))
for item in keywords:
    print(item[0],item[1])
print('-'*100)
```

```python
# 输出重要的句子
tr4s = TextRank4Sentence()
tr4s.analyze(text=text, lower=True, source = 'all_filters')
print('摘要：')
# 重要性较高的三个句子
for item in tr4s.get_key_sentences(num=5):
	# index是语句在文本中位置，weight表示权重
    print(item.index, item.weight, item.sentence)
```
