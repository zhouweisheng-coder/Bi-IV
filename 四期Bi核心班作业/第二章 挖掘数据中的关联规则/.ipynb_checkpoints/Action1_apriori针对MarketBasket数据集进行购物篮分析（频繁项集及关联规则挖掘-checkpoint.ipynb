# Apriori算法实战

```python
import pandas as pd
#数据加载
dataset = pd.read_csv('./Market_Basket_Optimisation.csv',header=None)
dataset.head()
```

```python
# 把数据存放在transactions
transactions = []
for i in range(0,dataset.shape[0]):
    temp = []
    for j in range(0,dataset.shape[1]):
        if str(dataset.values[i,j]) != 'nan':
            temp.append(str(dataset.values[i,j]))	
    # print(temp)
    transactions.append(temp)
# print(transactions)
```

## 采用efficient_apriori工具包

```python
from efficient_apriori import apriori
itemsets, rules = apriori(transactions,min_support=0.02,min_confidence=0.25)
```

```python
#频繁项集
itemsets
```

```python
#关联规则
rules
```

```python
data1 = pd.DataFrame(rules,columns=['关联规则'])
data1
```

```python
# 通过字符串拆分将关联规则列表str拆分成str和各指标数据，以方便查看及排序
Association_rules_list,confidence_list,support_list,lift_list,conv_list = [],[],[],[],[]
for key,value in data1['关联规则'].iteritems():
#     print(str(value))
    s = str(value)
    conf_index = s.find('conf')
    Association_rules = s[0:conf_index-2]
    Association_rules_list.append(Association_rules)
    
    conf = float(s[conf_index+6:conf_index+11])
    confidence_list.append(conf)
    
    support_index = s.find('supp')
    supp = float(s[support_index+6:support_index+11])
    support_list.append(supp)
    
    lift_index = s.find('lift')
    lift =float(s[lift_index+6:lift_index+11])
    lift_list.append(lift)
    
    conv_index = s.find('conv')
    conv = float(s[conv_index+6:conv_index+11])
    conv_list.append(conv)
```

```python
rules = pd.DataFrame({'关联规则':Association_rules_list,'confidence':confidence_list,'support':support_list,'lift':lift_list,'conv':conv_list})
rules = rules.sort_values(by="lift" , ascending=False) 
rules
```

## 采用mlxtend.frequent_patterns工具包

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
dataset.head()
```

```python
# 数据处理
transactions2 = []
for i in range(0,dataset.shape[0]):
    temp = []
    for j in range(0,dataset.shape[1]):
        if str(dataset.values[i,j]) != 'nan':
            temp.append(str(dataset.values[i,j]))
    temp='|'.join(temp)        
    # print(temp)
    transactions2.append(temp)
```

```python
data_Market_Basket = pd.DataFrame(transactions2,columns=['Market_Basket'])
print('数据处理后：')
data_Market_Basket.head(10)
```

```python
# one_hot编码
hot_encoded = data_Market_Basket.drop('Market_Basket',1).join(data_Market_Basket.Market_Basket.str.get_dummies('|'))
hot_encoded.head()
```

```python
# 挖掘频繁项集，最小支持度为0.04
itemsets2 = apriori(hot_encoded,use_colnames=True, min_support=0.02)
# 按照支持度从大到小排序
itemsets2 = itemsets2.sort_values(by="support" , ascending=False) 
print('-'*10, '频繁项集', '-'*10)
itemsets2
```

```python
# 根据频繁项集计算关联规则
rules2 =  association_rules(itemsets2, metric='lift',min_threshold=1)
# 按照提升度从大到小进行排序
rules2 = rules2.sort_values(by="lift" , ascending=False) 
print('-'*38, '关联规则', '-'*40)
rules2
```
