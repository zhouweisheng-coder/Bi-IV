# WordCloud实战

```python
# -*- coding:utf-8 -*-
from wordcloud import WordCloud
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
from lxml import etree
from nltk.tokenize import word_tokenize
```

```python
# 去掉停用词
def remove_stop_words(f):
    stop_words = ['nan']
    for stop_word in stop_words:
        f = f.replace(stop_word, '')
    return f

# 生成词云
def create_word_cloud(f):
    print('根据词频，开始生成词云!')
    f = remove_stop_words(f)
    cut_text = word_tokenize(f)
    #print(cut_text)
    cut_text = " ".join(cut_text)
    wc = WordCloud(
        max_words=100,
        width=2000,
        height=1200,
#         collocations=False,
    )
    wordcloud = wc.generate(cut_text)
    # 写词云图片
    wordcloud.to_file("wordcloud.jpg")
    # 显示词云文件
    plt.imshow(wordcloud)
    plt.axis("off")
    plt.show()
```

```python
# 数据加载
data = pd.read_csv('./Market_Basket_Optimisation.csv', header = None)
```

```python
all_word = ''
transactions = []
for i in range(0, data.shape[0]):
    transactions.append([str(data.values[i, j]) for j in range(0, 20) if str(data.values[i, j]) != 'nan'])
all_word = " ".join('%s' %id for id in transactions)
all_word[:1000]
```

```python
create_word_cloud(all_word)
```

# 统计词频

```python
# Counter统计词频
from collections import Counter
temp = []
for i in range(len(transactions)):
    for j in transactions[i]:
        temp.append(j)
dictionary = Counter(temp)
```

```python
dictionary.most_common(10)
```

```python
# Top10商品
df_top10 = pd.DataFrame(dictionary.most_common(10),columns=['商品名称','商品计数'],index=range(1,11))
df_top10
```
